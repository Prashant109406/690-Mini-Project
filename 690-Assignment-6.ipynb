# Mini Project
This dataset includes the Baltimore public school salaries information compiled by The Baltimore Sun Media Group from 2014.This data set includes the name and job title of the employees along with the Original hiring and Latest hiring dates and their earnings such as regular earnings,overtime and some additional earning over that period.The data was discovered from the link https://umbc.app.box.com/notes/375877568053?s=inwbinsctcmedcknp7h8r3tm1das5vs9

Objective: Extract file from the url and do exploratory data analysis on the columns such as to summarize the regular earnings as well as other earning columns.Also EDA can be done on Original and latest hiring dates.

Primary Observation
The primary observation about the dataset is that the data is about the salary and the extra earning earned by the employees. The data also tells about the orginal hiring date and also if the employee had left the company and rejoined, so the latest hiring date is also given.

```python
#importing libraries 
import urllib3
import certifi
import io
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. Getting the dataset

```python
#importing the dataset
http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED',ca_certs=certifi.where())
r = http.request('GET','http://salaries-archive.news.baltimoresun.com/baltimore-city-public-schools/updated-2015/cy2014-bcps-compensation-table-export.csv',preload_content=False)
salaries=pd.read_csv(r)
salaries
```

```python
#sorting the column vales as per Total Salary
formatted_df = salaries.sort_values(by=["Total Salary"])

formatted_df.head(100)
```

### 2. Checking the dataset for abmormilities

```python
formatted_df.isnull().sum()
```

```python
formatted_df.describe()
```

Salaries cannot be negative hence removing those values

```python
#if salary is less then zero then removing it
formatted_df[formatted_df["Total Salary"]<0]=0
formatted_df[formatted_df["Add Earn"]<0]=0
formatted_df[formatted_df["Reg Earn"]<0]=0
columns = ['Name']
formatted_df = formatted_df.replace(0, pd.np.nan).dropna(axis=0, how='any', subset=columns).fillna(0)
formatted_df
```

### 4.Checking for unique entries

```python
for x in formatted_df.columns:

    if (formatted_df[x].nunique()!=1):
        print(x,"no same value rows")
```

```python
def uniq_per_col(df,count):
    for this_column in df.columns:
        print("==== ",this_column,"has",df[this_column].nunique(),"unique entries ====")
        print(df[this_column].value_counts().head(count))
uniq_per_col(formatted_df,12)
```

### 5. Visualization

##### Visualization 1

```python
#plotting the no. of employees for that same job

from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
formatted_df.sample(10).groupby('Job')['Name'].nunique().plot(kind='bar')
plt.xlabel("Job Title")
plt.ylabel("no. of Employees")
plt.title("no. of employees for that same job")
plt.show()
```

##### Visualization 2

```python
#Converting Date in Datetime format

formatted_df['Original_Hire_as_date'] = pd.to_datetime(formatted_df['Original Hire'],infer_datetime_format=True)
formatted_df['Latest_Hire_as_date'] = pd.to_datetime(formatted_df['Latest Hire'],infer_datetime_format=True)

formatted_df=formatted_df.drop(['Original Hire','Latest Hire'],axis=1)
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 20), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(x='Original_Hire_as_date', y='Latest_Hire_as_date', data=formatted_df);
plt.xlabel("Original Hire Year")
plt.ylabel("Latest Hire Year")
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)

plt.title(label="Original Hiring vs Latest Hiring of Employees",fontsize=20)
plt.show();
```

The above scatter plot shows that most of the employess were rehired from the year 2000
